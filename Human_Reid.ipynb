```python
import torch 
from torch import nn
import torchreid
```

```python
dataManager = torchreid.data.ImageDataManager(
    root='data',
    sources='market1501',
    targets='market1501',
    height=256,
    width=128,
)
```

```python
model = torchreid.models.build_model(
    name='osnet_x1_0',
    num_classes=dataManager.num_train_pids,
    loss='softmax',
    pretrained=True
)
model = model.cuda()

optimizer = torchreid.optim.build_optimizer(
    model,
    optim='amsgrad',
    lr=0.005
)

scheduler = torchreid.optim.build_lr_scheduler(
    optimizer,
    lr_scheduler='single_step',
    stepsize=20
)
```

```python
engine = torchreid.engine.ImageSoftmaxEngine(
    dataManager,
    model,
    optimizer,
    scheduler
)

engine.run(
    save_dir='log/osnet_x1_0_market1501',
    max_epoch=30,
    eval_freq=10,
    print_freq=20,
    test_only=False
)
```

```python
engine.test(
    save_dir = 'log/visrank_osnet_x1_0',
    visrank = True,
    visrank_topk = 10,
    ranks = [1, 5, 10]
)
```
